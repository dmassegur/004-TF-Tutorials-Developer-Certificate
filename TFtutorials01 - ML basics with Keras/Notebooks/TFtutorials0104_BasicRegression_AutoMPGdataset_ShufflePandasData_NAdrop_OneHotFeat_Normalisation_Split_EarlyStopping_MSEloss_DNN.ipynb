```
# TensorFlow Tutorials
# ML basics with Keras 
# Basic Regression
# https://www.tensorflow.org/tutorials/keras/regression
# Auto MPG dataset: https://archive.ics.uci.edu/ml/datasets/auto+mpg
# TF Keras: https://www.tensorflow.org/guide/keras
# Keras get file: https://www.tensorflow.org/api_docs/python/tf/keras/utils/get_file
# pd map: reassign values in a series: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html
# pd get_dummies  to one-hot encode a categorical column: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html
# pd pop: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pop.html
# TF Keras EarlyStopping: https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping
# np expand dimension: https://docs.scipy.org/doc/numpy/reference/generated/numpy.expand_dims.html
# np collapse dimension: https://docs.scipy.org/doc/numpy/reference/generated/numpy.squeeze.html


# https://www.tensorflow.org/tutorials/structured_data/feature_columns
# https://www.tensorflow.org/tutorials/structured_data/imbalanced_data

# tfds.load: https://www.tensorflow.org/datasets/api_docs/python/tfds/load
# imdb tfds dataset: https://www.tensorflow.org/datasets/catalog/imdb_reviews
# TF Hub (Transfer Learning): https://www.tensorflow.org/hub

# TF Keras Embedding: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding
# TF Hub Keraslayer: https://www.tensorflow.org/hub/api_docs/python/hub/KerasLayer
# TF Keras regularizers: https://keras.io/api/layers/regularizers/    https://www.tensorflow.org/api_docs/python/tf/keras/regularizers/Regularizer
# TF Keras Cnv1D: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv1D
# TF Keras LSTM: https://www.tensorflow.org/api_docs/python/tf/keras/layers/LSTM
# TF Keras Dense: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense

# Live training plots: https://gist.github.com/stared/dfb4dfaf6d9a8501cd1cc8b8cb806d2e


# Use seaborn for pairplot
!pip install -q seaborn

# Use some functions from tensorflow_docs
!pip install -q git+https://github.com/tensorflow/docs

import pathlib

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


import tensorflow as tf
print(tf.__version__)
from tensorflow import keras
from tensorflow.keras import layers

from IPython import display
from IPython.display import clear_output
# from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from datetime import *

import tensorflow_docs as tfdocs
import tensorflow_docs.plots
import tensorflow_docs.modeling

print("TF Version: ", tf.__version__)
# print("Eager mode: ", tf.executing_eagerly())
# print("Hub version: ", hub.__version__)
print("GPU is", "available" if tf.config.experimental.list_physical_devices("GPU") else "NOT AVAILABLE")

print('\n')


starttime = datetime.now()


# Download the dataset:
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin']
raw_dataset = pd.read_csv(dataset_path, sep=' ', names= column_names, na_values= "?", comment= '\t', skipinitialspace= True, header= None)
print(raw_dataset.columns)
print(raw_dataset.shape)
dataset = raw_dataset.copy()

# Shuffle the data:
dataset = dataset.reindex(np.random.permutation(dataset.index))
print(dataset.head())


print('\n')


# Clean the data 
# The dataset contains a few unknown values. 
print(dataset.isna().sum()) 

# Drop non-available values:
dataset = dataset.dropna()
print(dataset.isna().sum()) 


print('\n')


# The "Origin" column is really categorical, not numeric. So convert that to a one-hot:
# we use pd map to re-assign series values in a series.  Map values of Series according to input correspondence. https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html
dataset['Origin'] = dataset['Origin'].map({1: 'USA', 2: 'Europe', 3: 'Japan'})
print(dataset.tail())
print(dataset.head())
# Convert categorical variable into dummy/indicator variables. https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html
dataset = pd.get_dummies(dataset, prefix='Origin', prefix_sep='_')
print(dataset.tail())


print('\n')


# Inspect the data
# Have a quick look at the joint distribution of a few pairs of columns from the training set.
sns.pairplot(dataset[["MPG", "Cylinders", "Displacement", "Weight"]], diag_kind="kde")
display.display(dataset.corr(method ='pearson'))  # correlation matrix
display.display(dataset.describe())


print('\n')


# Normalising the data:
# we don't normalize the label MPG (1st column)!
# It is good practice to normalize features that use different scales and ranges. Although the model might converge without feature normalization, it makes training more difficult, and it makes the resulting model dependent on the choice of units used in the input.
dataset.iloc[:, 1:] = (dataset.iloc[:, 1:] - dataset.iloc[:, 1:].mean()) / dataset.iloc[:, 1:].std()
print(dataset.head())
display.display(dataset.describe())


print('\n')


# Split the data:
train_split = 0.8
train_split = int(train_split * dataset.shape[0])
train_data = dataset.iloc[:train_split, :]
test_data = dataset.iloc[train_split:, :]

# alternative
# train_dataset = dataset.sample(frac=0.8,random_state=0)
# test_dataset = dataset.drop(train_dataset.index)

print(train_data.shape)
print(train_data.head())
print(test_data.shape)
print(test_data.head())


print('\n')


# Split into features and labels:
# pd pop  Return item and drop from frame:   https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pop.html
train_labels = train_data.pop('MPG')  # drops the 'mpg' column from train_data and adds it to train_labels
test_labels = test_data.pop('MPG')  # drops the 'mpg' column from test_data and adds it to test_labels


print('\n')


# Converting datasets to numpy:
train_data = train_data.values
train_labels = train_labels.values
test_data = test_data.values
test_labels = test_labels.values

print(train_data.shape)
print(train_labels.shape)
print(test_data.shape)
print(test_labels.shape)


et1 = datetime.now() - starttime

print( '\n-> Elapsed execution time: %0.4f seconds.\n' %(et1.total_seconds()) )
```

## Build the model


```
# Build a train model:

def build_model():
  model = tf.keras.Sequential()

  model.add(tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l1_l2(l1=0.01, l2=0.01), input_shape=[train_data.shape[1]]))   # input_shape=(train_data.shape[1],) works too
  # model.add(tf.keras.layers.Dropout(0.5))
  model.add(tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l1_l2(l1=0.01, l2=0.01)))
  # model.add(tf.keras.layers.Dropout(0.5))

  model.add(tf.keras.layers.Dense(1))  # this is regression, no activation needed on output


  from tensorflow.keras.optimizers import Adam
  from tensorflow.keras.optimizers import RMSprop
  from tensorflow.keras.optimizers import SGD
  model.compile(
      optimizer=Adam(lr=0.0003),
      loss = 'mse',
      metrics = ['accuracy','mse','mae']
  )


  return model


model = build_model()

model.summary()
```

```
starttime = datetime.now()

# Clearing the session
tf.keras.backend.clear_session()
tf.random.set_seed(51)
np.random.seed(51)


# Callback for training target:
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        target=0.9999
        if logs['accuracy']>target :
            print('\nReached %0.2f%% training accuracy. Training converged!' %(target*100))
            self.model.stop_training = True
        
            
callback = myCallback()


# Callback for Live Training Plots:
class PlotLosses(tf.keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        self.accu = []
        self.val_accu = []
        self.mse = []
        self.val_mse = []
        self.mae = []
        self.val_mae = []

        self.fig = plt.figure(figsize=(15,6))
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        # self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.accu.append(logs.get('accuracy'))
        self.val_accu.append(logs.get('val_accuracy'))
        self.mse.append(logs.get('mse'))
        self.val_mse.append(logs.get('val_mse'))
        self.mae.append(logs.get('mae'))
        self.val_mae.append(logs.get('val_mae'))
        self.i += 1
        
        clear_output(wait=True)
        
        plt.subplot(2,2,1)
        plt.plot(self.x, self.accu, 'b', label="accuracy")
        plt.plot(self.x, self.val_accu, 'r', label="val_accuracy")
        plt.grid()
        plt.xlabel('epochs')
        plt.ylabel('accuracy')
        plt.legend()
        
        plt.subplot(2,2,2)
        plt.plot(self.x, self.losses, 'b', label="loss")
        plt.plot(self.x, self.val_losses, 'r', label="val_loss")
        plt.grid()
        plt.xlabel('epochs')
        plt.ylabel('loss')
        plt.legend()

        plt.subplot(2,2,3)
        plt.plot(self.x, self.mse, 'b', label="mse")
        plt.plot(self.x, self.val_mse, 'r', label="val_mse")
        plt.grid()
        plt.xlabel('epochs')
        plt.ylabel('mse')
        plt.legend()

        plt.subplot(2,2,4)
        plt.plot(self.x, self.mae, 'b', label="mae")
        plt.plot(self.x, self.val_mae, 'r', label="val_mae")
        plt.grid()
        plt.xlabel('epochs')
        plt.ylabel('mae')
        plt.legend()

        plt.show();


plot_losses = PlotLosses()


# Train the model:
history = model.fit(
    train_data,
    train_labels,
    # batch_size=16,
    epochs= 150,
    validation_split = 0.2,  # validation_data= (test_data, test_labels),
    callbacks= [callback, plot_losses, tfdocs.modeling.EpochDots()],    # The training for this tutorial runs for many short epochs. To reduce the logging noise use the tfdocs.EpochDots which simply prints a . for each epoch, and a full set of metrics every 100 epochs.
    verbose= 1
    )



et1 = datetime.now() - starttime

print( '\n-> Elapsed execution time: %0.4f seconds.\n' %(et1.total_seconds()) )
```

```
starttime = datetime.now()

# Plot the training curves:
print(history.history.keys())
loss = history.history['loss']
accu = history.history['accuracy']
mse = history.history['mse']
mae = history.history['mae']
val_loss = history.history['val_loss']
val_accu = history.history['val_accuracy']
val_mse = history.history['val_mse']
val_mae = history.history['val_mae']

plt.figure(figsize=(15,6))

plt.subplot(2,2,1)
plt.plot(accu, 'b', label='training accuracy')
plt.plot(val_accu, 'r', label='validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(loss, 'b', label='training loss')
plt.plot(val_loss, 'r', label='validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(mse, 'b', label='training mse')
plt.plot(val_mse, 'r', label='validation mse')
plt.xlabel('epochs')
plt.ylabel('mse')
plt.legend()
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(mae, 'b', label='training mae')
plt.plot(val_mae, 'r', label='validation mae')
plt.xlabel('epochs')
plt.ylabel('mae')
plt.legend()
plt.grid(True)

plt.show()



print('\n')


# Visualize the model's training progress using the stats stored in the history object
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
print(hist.tail())


print('\n')


plt.figure(figsize=(15,6))

plotter = tfdocs.plots.HistoryPlotter(smoothing_std=2)

plt.subplot(1,2,1)
plotter.plot({'Basic': history}, metric = "mae")
plt.ylim([0, 10])
plt.ylabel('MAE [MPG]')

plt.subplot(1,2,2)
plotter.plot({'Basic': history}, metric = "mse")
plt.ylim([0, 20])
plt.ylabel('MSE [MPG^2]')

plt.show()


print('\n')


# Evaluate accuracy: compare how the model performs on the test dataset:
print(model.evaluate(test_data, test_labels, verbose=1))
test_loss, test_accu, test_mse, test_mae = model.evaluate(test_data, test_labels, verbose=1)
print('Test accuracy: %0.2f%%' %(test_accu*100))
print('Test loss: %0.2f' %(test_loss))
print('Test mse: %0.2f' %(test_mse))
print('Test mae: %0.2f' %(test_mae))

results = model.evaluate(test_data, test_labels, verbose=1)
for name, value in zip(model.metrics_names, results):
  print('%s: %0.3f' %(name,value))


print('\n')


# Make predictions:
# predicting the labels for the test set:
predictions = model.predict(test_data)
# prediction on 10 test examples:
print(predictions[:10])

i = 0
for lab, pred in zip(test_labels, predictions):
  if i<10 :
    print('test example %d, label: %0.2f vs prediction: %0.2f' %(i,lab,pred))  # (pred>0.5).astype('int32')
  else :
    break
  i+=1


print('\n')


# Plotting predictions:
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.axes(aspect='equal')
plt.scatter(test_labels, predictions)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
plt.plot(lims, lims)
plt.grid()
# plt.show()

# Let's take a look at the error distribution.
plt.figure(figsize=(15,6))
plt.subplot(1,2,2)
test_labels = np.squeeze(test_labels)  # squeezes any array, matrix, tensor into a 1-dim array.
test_labels = np.expand_dims(test_labels, axis=1)
error = predictions - test_labels
print(test_labels.shape)
print(predictions.shape)
print(error.shape)
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [MPG]")
plt.ylabel("Count")
plt.grid()
plt.show()




et1 = datetime.now() - starttime

print( '\n-> Elapsed execution time: %0.4f seconds.\n' %(et1.total_seconds()) )
```

## Early Stopping:

This graph shows little improvement, or even degradation in the validation error after about 100 epochs. Let's update the `model.fit` call to automatically stop training when the validation score doesn't improve. We'll use an *EarlyStopping callback* that tests a training condition for  every epoch. If a set amount of epochs elapses without showing improvement, then automatically stop the training.

You can learn more about this callback [here](https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping).

```
# Clearing the session
tf.keras.backend.clear_session()
tf.random.set_seed(51)
np.random.seed(51)


model = build_model()

# The patience parameter is the amount of epochs to check for improvement
# TF Keras EarlyStopping: https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)


history = model.fit(
    train_data,
    train_labels,
    # batch_size=16,
    epochs= 300,
    validation_split = 0.2,  # validation_data= (test_data, test_labels),
    callbacks= [early_stop, callback, plot_losses, tfdocs.modeling.EpochDots()],    # The training for this tutorial runs for many short epochs. To reduce the logging noise use the tfdocs.EpochDots which simply prints a . for each epoch, and a full set of metrics every 100 epochs.
    verbose= 1
    )
```

## Conclusion

This notebook introduced a few techniques to handle a regression problem.

* Mean Squared Error (MSE) is a common loss function used for regression problems (different loss functions are used for classification problems).
* Similarly, evaluation metrics used for regression differ from classification. A common regression metric is Mean Absolute Error (MAE).
* When numeric input data features have values with different ranges, each feature should be scaled independently to the same range.
* If there is not much training data, one technique is to prefer a small network with few hidden layers to avoid overfitting.
* Early stopping is a useful technique to prevent overfitting.
